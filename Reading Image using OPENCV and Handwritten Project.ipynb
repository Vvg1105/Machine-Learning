```python
arr_data=[]
```

```python
y_va = []
```

```python
for fil in range(0,4):
    dirt = os.listdir(working_directory + '/data/training-images/'+str(fil))
    for pil in dirt:
        if(pil[-3:]=='png'):
            print(pil)
            path = working_directory + '/data/training-images/' + str(fil) + '/' + pil
            y_va.append(fil)
            img = cv2.imread(path, cv2.IMREAD_COLOR)
            new_img = img.reshape((img.shape[0]*img.shape[1]), img.shape[2])
            new_img = new_img.transpose()
            old_shape = new_img.shape
            arr = np.reshape(new_img, old_shape[0] * old_shape[1])
            arr_data.append(arr)
```

```python
arr_data
```

```python
arr_data
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()
```

```python
df
```

```python
for pos in range(len(arr_data)):
    df[pos]=arr_data[pos]
```

```python
df = df.T
```

```python
df
```

```python
df['Y Value'] = y_va
```

```python
df
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
pipeline=Pipeline([('classifier',KNeighborsClassifier(metric='euclidean'))])
```

```python
search_space = [{'classifier': [KNeighborsClassifier()],
                 'classifier__n_neighbors': np.arange(1, 25),
                 'classifier__metric': ['euclidean','manhattan','minkowski']}]
```

```python
clf = GridSearchCV(pipeline, search_space, cv=5)
```

```python
df = df.drop(columns = ['Y Value'], axis =1)
```

```python
df
```

```python
clf.fit(df,y_va)
```

```python
df_grid = pd.DataFrame(data=clf.cv_results_)
```

```python
df_grid
```
