```python
import pandas as pd
```

```python
df = pd.read_csv('adult.data',header=None,names=["Age", "Workclass", "fnlwgt", "Education", "Education num", "Marital Status", "Occupation","Relationship", "Race", "Sex", "Capital gain","Capital loss", "Hours per week", "Native country","Income level"])
```

```python
df
```

```python
arr_cat = []
```

```python
cols = df.columns[:-1]
```

```python
cols
```

```python
for pos in range(0,len(cols)):
    if(df[cols[pos]].dtype != 'int64'):
        arr_cat.append(cols[pos])
```

```python
arr_cat
```

```python
arr_no_cat = []
for pos in df.columns:
    if(pos not in arr_cat and pos!=df.columns[-1]):
        arr_no_cat.append(pos)
```

```python
arr_no_cat
```

```python
arr_encode=[]
for pos in arr_cat:
    arr_encode.append(pos+' Encode')
for pos in arr_no_cat:
    arr_encode.append(pos)
```

```python
df_pipe = pd.DataFrame(data=[],columns=arr_encode)
```

```python
df_pipe
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder_y = LabelEncoder()
```

```python
data_y = label_encoder_y.fit_transform(df[df.columns[-1]])
```

```python
label_encoder_y
```

```python
label = pd.DataFrame(data=data_y, columns=[df.columns[-1]+' Encode'])
```

```python
label
```

```python
df_pipe
```

```python
label_object = {}
for col in arr_cat:
    labelencoder = LabelEncoder()
    labelencoder.fit(df[col])
    df_pipe[col+' Encode'] = labelencoder.fit_transform(df[col])
    label_object[col] = labelencoder
print(label_object)
```

```python
df
```

```python
df_pipe
```

```python
for col in arr_no_cat:
    df_pipe[col] = df[col]
```

```python
df_pipe
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.pipeline import Pipeline
```

```python
X_train,X_test,y_train,y_test = train_test_split(df_pipe, label, test_size=0.30)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
pipeline_dt=Pipeline([('classifier',DecisionTreeClassifier(criterion='entropy'))])
```

```python
pipeline_randomforest=Pipeline([('classifier',RandomForestClassifier(criterion='entropy'))])
```

```python
pipelines = [pipeline_dt, pipeline_randomforest]
```

```python
pipelines[0]
```

```python
best_accuracy=0.0
```

```python
best_classifier=0
```

```python
best_pipeline = ''
```

```python
pipe_dict = {
    0: 'Decision Tree',
    1: 'RandomForest'
}
```

```python
for pipe in pipelines:
    pipe.fit(X_train, y_train)
```

```python
for i,model in enumerate(pipelines):
    print("{} Test Accuracy: {}".format(pipe_dict[i],model.score(X_test,y_test)))
```

```python
for i,model in enumerate(pipelines):
    if(model.score(X_test,y_test)<best_accuracy):
        best_accuracy = model.score(X_test,y_test)
        best_pipeline = model
        best_classifier=i
print("Classifier with best accuracy: {}".format(pipe_dict[best_classifier]))
```
