```python
import pandas as pd
```

```python
df = pd.read_csv('adult.data',header=None,names=["Age", "Workclass", "fnlwgt", "Education", "Education num", "Marital Status", "Occupation","Relationship", "Race", "Sex", "Capital gain","Capital loss", "Hours per week", "Native country","Income level"])
```

```python
df
```

```python
arr_cat = []
```

```python
cols = df.columns[:-1]
```

```python
cols
```

```python
for pos in range(0,len(cols)):
    if(df[cols[pos]].dtype != 'int64'):
        arr_cat.append(cols[pos])
```

```python
arr_cat
```

```python
arr_cat_encode=[]
for pos in arr_cat:
    arr_cat_encode.append(pos+' Encode')
```

```python
arr_cat_encode
```

```python
df_cat = pd.DataFrame(data=[],columns=arr_cat_encode)
```

```python
df_cat
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder_y = LabelEncoder()
```

```python
data_y = label_encoder_y.fit_transform(df[df.columns[-1]])
```

```python
label_encoder_y
```

```python
label = pd.DataFrame(data=data_y, columns=[df.columns[-1]+' Encode'])
```

```python
label
```

```python
data_y
```

```python
label_object = {}
for col in arr_cat:
    labelencoder = LabelEncoder()
    labelencoder.fit(df[col])
    df_cat[col+' Encode'] = labelencoder.fit_transform(df[col])
    label_object[col] = labelencoder
print(label_object)
```

```python
df_cat
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
arr_ns = []
for num in range(90,101):
    arr_ns.append(num)
```

```python
arr_params = []
for num in range(1,30):
    arr_params.append(num)
```

```python
paramters = {
    'min_samples_split': arr_params,
    'max_depth':arr_params,
    'n_estimators':arr_ns
}
```

```python
model = RandomForestClassifier(criterion = 'entropy')
```

```python
model = model.fit(df_cat,data_y)
```

```python
grid_search = GridSearchCV(model,paramters)
```

```python
grid_GBR = grid_search.fit(df_cat,data_y)
```

```python
print("The best score across ALL searched params:\n",grid_GBR.best_score_)
```

```python
print("The best parameters across ALL searched params:\n",grid_GBR.best_params_)
```

```python
df_grid_samples = pd.DataFrame(data = grid_GBR.cv_results_)
```
