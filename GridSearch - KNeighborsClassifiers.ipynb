```python
import pandas as pd
```

```python
df = pd.read_csv('adult.data',header=None,names=["Age", "Workclass", "fnlwgt", "Education", "Education num", "Marital Status", "Occupation","Relationship", "Race", "Sex", "Capital gain","Capital loss", "Hours per week", "Native country","Income level"])
```

```python
df
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
arr_no_cat=[]
```

```python
for pos in df.columns:
    if(df[pos].dtypes=='int64'):
        arr_no_cat.append(pos)
```

```python
print(arr_no_cat)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder_y = LabelEncoder()
```

```python
label = label_encoder_y.fit_transform(df[df.columns[-1]])
```

```python
label
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
arr_params = []
for num in range(29,100):
    arr_params.append(num)
```

```python
paramters = {
    'n_neighbors': arr_params
}
```

```python
model = KNeighborsClassifier()
```

```python
grid_search = GridSearchCV(model,paramters)
```

```python
df_no_cat = pd.DataFrame(data=[],columns=arr_no_cat)
```

```python
for pos in df_no_cat.columns:
    df_no_cat[pos] = df[pos]
```

```python
df_no_cat
```

```python
grid_GBR = grid_search.fit(df_no_cat,label)
```

```python
print("The best score across ALL searched params:\n",grid_GBR.best_score_)
```

```python
print("The best parameters across ALL searched params:\n",grid_GBR.best_params_)
```

```python
df_grid = pd.DataFrame(data = grid_GBR.cv_results_)
```

```python
df_grid
```
