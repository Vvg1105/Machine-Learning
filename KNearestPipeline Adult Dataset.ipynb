```python
import pandas as pd
```

```python
df = pd.read_csv('adult.data',header=None,names=["Age", "Workclass", "fnlwgt", "Education", "Education num", "Marital Status", "Occupation","Relationship", "Race", "Sex", "Capital gain","Capital loss", "Hours per week", "Native country","Income level"])
```

```python
df
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
arr_no_cat=[]
```

```python
for pos in df.columns:
    if(df[pos].dtypes=='int64'):
        arr_no_cat.append(pos)
```

```python
print(arr_no_cat)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder_y = LabelEncoder()
```

```python
label = label_encoder_y.fit_transform(df[df.columns[-1]])
```

```python
label
```

```python
from sklearn.pipeline import Pipeline
```

```python
df_no_cat = pd.DataFrame(data=[],columns=arr_no_cat)
```

```python
for pos in df_no_cat.columns:
    df_no_cat[pos] = df[pos]
```

```python
df_no_cat
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
data = scaler.fit_transform(df_no_cat)
```

```python
df_n = pd.DataFrame(data,columns = arr_no_cat)
```

```python
df_n
```

```python
pipeline=Pipeline([('classifier',KNeighborsClassifier(metric='euclidean'))])
```

```python
arr_params = []
for num in range(29,51):
    arr_params.append(num)
```

```python
search_space = [{'classifier': [KNeighborsClassifier()],
                'classifier__n_neighbors': arr_params,
                'classifier__metric':['euclidean']},
               {'classifier': [KNeighborsClassifier()],
                'classifier__n_neighbors': arr_params,
                'classifier__metric':['minkowski']},
               {'classifier': [KNeighborsClassifier()],
                 'classifier__n_neighbors': arr_params,
                  'classifier__metric':['manhattan']},]
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf = GridSearchCV(pipeline, search_space)
```

```python
clf
```

```python
clf.fit(df_n,label)
```

```python
df_grid = pd.DataFrame(data=clf.cv_results_)
```

```python
df_grid
```

```python
df_grid[df_grid['rank_test_score']<=5]
```

```python
df_grid[df_grid['param_classifier__metric']=='euclidean']
```

```python
pipeline_scale = Pipeline([('scale', StandardScaler()),
    ('classifier',KNeighborsClassifier(metric='euclidean'))])
```

```python
models = GridSearchCV(pipeline_scale, search_space)
```

```python
models.fit(df_no_cat,label)
```

```python
df_grid_se = pd.DataFrame(data=models.cv_results_)
```
