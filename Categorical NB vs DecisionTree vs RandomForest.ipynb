```python
import pandas as pd
```

```python
df = pd.read_csv('adult.data',header=None,names=["Age", "Workclass", "fnlwgt", "Education", "Education num", "Marital Status", "Occupation","Relationship", "Race", "Sex", "Capital gain","Capital loss", "Hours per week", "Native country","Income level"])
```

```python
df
```

```python
arr_cat = []
```

```python
cols = df.columns[:-1]
```

```python
cols
```

```python
for pos in range(0,len(cols)):
    if(df[cols[pos]].dtype != 'int64'):
        arr_cat.append(cols[pos])
```

```python
arr_cat
```

```python
arr_encode=[]
for pos in arr_cat:
    arr_encode.append(pos+' Encode')
```

```python
df_pipe = pd.DataFrame(data=[],columns=arr_encode)
```

```python
df_pipe
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder_y = LabelEncoder()
```

```python
data_y = label_encoder_y.fit_transform(df[df.columns[-1]])
```

```python
label_encoder_y
```

```python
label = pd.DataFrame(data=data_y, columns=[df.columns[-1]+' Encode'])
```

```python
label
```

```python
arr_y = []
for pos in label[df.columns[-1]+' Encode']:
    arr_y.append(pos)
```

```python
label_object = {}
for col in arr_cat:
    labelencoder = LabelEncoder()
    labelencoder.fit(df[col])
    df_pipe[col+' Encode'] = labelencoder.fit_transform(df[col])
    label_object[col] = labelencoder
print(label_object)
```

```python
df_pipe
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.naive_bayes import CategoricalNB
```

```python
pipeline=Pipeline([('classifier',DecisionTreeClassifier(criterion='entropy'))])
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
search_space = [{'classifier': [RandomForestClassifier(criterion='entropy')],
                 'classifier__n_estimators': [10, 100, 1000],
                 'classifier__min_samples_split': [2, 3]},
               {'classifier': [DecisionTreeClassifier(criterion='entropy')],
                 'classifier__max_depth': [10, 20, 40],
                 'classifier__min_samples_split': [2, 3]},
               {'classifier': [CategoricalNB()],
                 'classifier__alpha': [1, 2]}]
```

```python
clf = GridSearchCV(pipeline, search_space)
```

```python
clf
```

```python
best_model = clf.fit(df_pipe, arr_y)
```

```python
best_model.best_estimator_.get_params()['classifier']
```

```python
df_grid = pd.DataFrame(data=clf.cv_results_)
```
